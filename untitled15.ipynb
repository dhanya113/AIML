```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/breast-cancer (6).csv")
print(df)
print(df.info())
print(df.isnull().sum())
print(df.describe())
print(df.columns)
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['diagnosis']=le.fit_transform(df['diagnosis'])
print(df['diagnosis'])
x=df.drop('diagnosis',axis=1)
print(x)
y=df['diagnosis']
print(y)
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.23,random_state=0)
print("xtrain shape:",xtrain.shape)
print("xtest shape:",xtest.shape)
print("ytrain shape:",ytrain.shape)
print("ytest shape:",ytest.shape)
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(xtrain,ytrain)
train_accuracy=model.score(xtrain,ytrain)
test_accuracy=model.score(xtest,ytest)
print('train accuracy:',train_accuracy)
print('test accuracy:',test_accuracy)
train_prediction=model.predict(xtrain)
print(train_prediction)
test_prediction=model.predict(xtest)
print(test_prediction)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, mean_absolute_error
df=pd.read_csv('/content/housing.csv')
print(df)
df.info()
print(df.isnull().sum())
print(df.head())
plt.scatter(df.built_up_area,df.rent,color='red',marker='*')
plt.xlabel('built_up_area')
plt.ylabel('rent')
plt.title('Scatter plot')
plt.show()
x=df.drop('rent',axis=1)
print(x)
y=df.rent
print(y)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
print("xtrain shape:",x_train.shape)
print("xtest shape:",x_test.shape)
print("ytrain shape:",y_train.shape)
print("ytest shape:",y_test.shape)
reg=linear_model.LinearRegression()
reg.fit(x,y)
y_pred=reg.predict(x_test)
reg.predict([[650]])
print(reg.coef_)
print(reg.intercept_)
z=9.7900215*650+1128.879253617697
print(z)
train_score=reg.score(x_train,y_train)
test_score=reg.score(x_test,y_test)
print('train score(R-squared):',train_score)
print('test score(R-squared):',test_score)
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
print('mean_squared_error:',mse)
print('mean_absolute_error:',mae)
```
